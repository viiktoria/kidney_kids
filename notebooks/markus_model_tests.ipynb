```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

from kidney_kids.data import *
```

```python
X_train, X_test, y_train, y_test = get_cleaned_data()
X_preproc = get_preproc_data(X_train)
y_train
```

```python
# use with X_train and the algortihm: 'tree' or 'knn', by default: LogRegrerssion!

def GridS_tree(X, y):
    
    model = DecisionTreeClassifier()
   
    params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [3,7,15],
    'min_samples_split' : [2, 3, 6]
    }
        

    search = GridSearchCV(model, param_grid=params , scoring='recall', n_jobs=-1, cv= 4, verbose= 2.1)
    result = search.fit(X,y)

    results_df = pd.DataFrame(result.cv_results_)

    return [result.best_estimator_, results_df, result.best_params_]
```

```python
%%time
GridS_tree(X_preproc,y_train)
```

```python
def GridS_LogR(X, y):
    
    model = LogisticRegression()
    

    params ={
    'solver': ['newton-cg', 'saga', 'liblinear'],
    'penalty': ['none', 'l2', 'l1', 'elastcinet'],
    'C': [1e-100, 1e-4, 1e-3]}

    search = GridSearchCV(model, param_grid=params, scoring='recall', n_jobs=-1, cv= 5)
    result = search.fit(X,y)

    results_df = pd.DataFrame(result.cv_results_)

    return [result.best_estimator_, results_df, result.best_params_]
```

```python
GridS_LogR(X_preproc, y_train)
```

```python
def GridS_KNN(X, y):
    
    model = KNeighborsClassifier()
    params = {
    'n_neighbors': np.arange(5,15),
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'metric': ['euclidean', 'manhattan', 'chebyshev', 'minkowski'], #there are more metrices
    'weights': ['uniform', 'distance']
    }

    search = GridSearchCV(model, param_grid=params, scoring='recall', n_jobs=-1, cv= 10)
    result = search.fit(X,y)

    results_df = pd.DataFrame(result.cv_results_)

    return [result.best_estimator_, results_df, result.best_params_]
```

```python
%%time
GridS_KNN(X_preproc, y_train)
```
